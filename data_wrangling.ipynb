```python
# import all the things
import sys, pandas as pd, numpy as np, json, random, re, gzip, textwrap
```

# Data

```python
# load all the dataframes
gutenberg = pd.read_csv("/content/drive/Shareddrives/Kaneb Center Course Design Academy/Corpus_Adventures/Data/gutenberg_combined.csv")
pFound = pd.read_csv("/content/drive/Shareddrives/Kaneb Center Course Design Academy/Corpus_Adventures/Data/pFound.csv")
poetsOrg = pd.read_csv("/content/drive/Shareddrives/Kaneb Center Course Design Academy/Corpus_Adventures/Data/poetsOrg.csv")
poemDB = pd.read_csv("/content/drive/Shareddrives/Kaneb Center Course Design Academy/Corpus_Adventures/Data/poemDB.csv")
gt = pd.read_csv("/content/drive/Shareddrives/Kaneb Center Course Design Academy/Notebooks/gt.csv")
```

```python
# show list of columns
#gutenberg.info()

# create subset with select columns
gSubset = gutenberg[['poem', 'title', 'authors']].copy()

# add column
gSubset['source'] = "Project Gutenberg"

# rename columns
columns = ['text', 'title', 'author', 'source']
gSubset.columns = columns

# reorder columns
order = ['title', 'text', 'author', 'source']
gSubset = gSubset[order]
gSubset
```

```python
# show list of columns
# pFound.info()
pFoundSubset = pFound[['Title', 'Poem', 'Poet']].copy()
pFoundSubset['source'] = "Poetry Foundation"
columns = ['title', 'text', 'author', 'source']
pFoundSubset.columns = columns
pFoundSubset.head()
```

```python
# show list of columns
#poetsOrg.info()
pOrgSubset = poetsOrg[['author', 'attribution', 'title', 'text']].copy()
pOrgSubset['source'] = 'Poets.Org'
order = ['title', 'text', 'author', 'attribution', 'source']
pOrgSubset = pOrgSubset[order]
pOrgSubset.head()
```

```python
# show column info
# poemDB.info()
poemDBSubset = poemDB[['title', 'author', 'text']].copy()
poemDBSubset['source'] = "PoemDB"
order = ['title', 'text', 'author', 'source']
poemDBSubset = poemDBSubset[order]
poemDBSubset.head()
```

```python
# show column info
# gt.info()
gtSubset = gt[['title', 'author', 'text']].copy()
gtSubset['source'] = 'Golden Treasury'
order = ['title', 'text', 'author', 'source']
gtSubset = gtSubset[order]
gtSubset.head()
```

# Filtering

```python
# function to filter out regex
def regexFilter(df):
  df['title'] = df['title'].str.replace(r"\r\r", '')
  df['title'] = df['title'].str.replace(r"\n", '')
  df['text'] = df['text'].str.replace(r"\r\r", '')
  return df

# function calls
regexFilter(gtSubset)
regexFilter(poemDBSubset)
regexFilter(gSubset)
regexFilter(pFoundSubset)
regexFilter(pOrgSubset)
```

```python
# function to get line count
def lineCount(df):
  value = df['text'].str.count('\n')
  df['lineCount'] = (value+1)
  return df

lineCount(gtSubset)
lineCount(poemDBSubset)
lineCount(gSubset)
lineCount(pFoundSubset)
lineCount(pOrgSubset)
```

```python
test = pOrgSubset['text'].str.split("\n")

avg = []

for row in test:
  characters = 1
  try:
    for r in row:
      characters += len(r)
    print(characters)
  except:
    continue
  print(characters/len(row))
  
```

```python
# function to get average number of characters per line
def avgChar(df):
  test = df['text'].str.split("\n")
  avg = []
  for row in test:
    characters = 1
    try:
      for r in row:
        characters += len(r)
    except:
      pass
    try:
      avg.append(characters/len(row))
    except:
      avg.append(0)
   
  df['avgChar'] = avg
  return df


# function calls
avgChar(gtSubset)
avgChar(poemDBSubset)
avgChar(gSubset)
avgChar(pFoundSubset)
avgChar(pOrgSubset)
```

# Merging Experiments

```python
# concatenate based on common structure
combined = pd.concat([gtSubset, poemDBSubset, gSubset, pFoundSubset, pOrgSubset], axis=0)
combined.to_csv("combined_subset.csv", index=False)
combined
```
